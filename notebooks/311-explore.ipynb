```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df311 = pd.read_csv(
    "../data/in/311_Service_Requests_20240307.csv",
    parse_dates=['Open Date','Closed Date'],
    date_format='%m/%d/%Y %I:%M:%S %p'
)
```

```python
df311['days_open']=(df311['Closed Date'] - df311['Open Date']).dt.seconds
```

```python
sns.histplot(df311['days_open'], kde=True)
plt.show()
```
